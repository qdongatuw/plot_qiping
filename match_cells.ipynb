```python
import pandas as pd
```

```python
f_cells = r'C:\Users\dongq\OneDrive\mouse patch-seq\result\cell list_mecp2.xlsx'
raw = r'C:\Users\dongq\OneDrive\mouse patch-seq\result\2023-05-26-19-34.csv'

```

```python
df_list = pd.read_excel(f_cells)
df_raw = pd.read_csv(raw)
```

```python
df_list
```

```python
data_dict = {}

# DataFrame
for index, row in df_list.iterrows():
    # ID
    sample_id = row.iloc[0]
    age = row.iloc[2]
    genotype = row.iloc[3]
    results = row.iloc[4:].tolist()  # 
    data_dict[sample_id] = dict()

    data_dict[sample_id]['age'] = age
    data_dict[sample_id]['genotype'] = genotype
    data_dict[sample_id]['files'] = []
    # ID
    results_filtered = [result for result in results if pd.notna(result)]
    if sample_id in data_dict:
        data_dict[sample_id]['files'].extend(results_filtered)
    
```

```python
raw_dict = df_raw.to_dict(orient='records')
```

```python
str(raw_dict[0]['-1000.0 pA']) == 'nan'
```

```python
raw_dict_key_file = dict()
for record in raw_dict:
    raw_dict_key_file[record['File']] = record
    
```

```python
summary_dict = dict()
for cell in data_dict:
    summary_dict[cell] = dict()
    files = data_dict[cell]['files']
    for k in files:
        file = k + '.abf'
        if file in raw_dict_key_file:
            result = raw_dict_key_file[file]
            for i in result:
                if str(result[i]) != 'nan' and i not in summary_dict[cell]:

                    summary_dict[cell][i] = result[i]
```

```python
df_summary = pd.DataFrame(summary_dict)
```

```python
df_summary.T.to_csv('summary.csv')
```
